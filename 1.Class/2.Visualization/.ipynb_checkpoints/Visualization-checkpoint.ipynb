```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

Handling Null Values

```python
df.isnull().sum()
```

```python
null_mean=df.isnull().mean()*100
print(null_mean)
```

```python
df.drop('Cabin', axis=1, inplace=True)
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mode(), inplace=True)
```

```python
df.head()
```

Visualization

```python
val=df["Pclass"].value_counts() #Counts unique values in the "Pclass" column
print(val)
```

Bar Graph

```python
plt.figure(figsize=(8,6))
val.plot(kind='bar', color='purple', edgecolor='black')
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.xticks(rotation = 0)
plt.show()
```

Histogram

```python
plt.figure(figsize=(8,6))
plt.hist(df['Age'], bins=24, color='orange', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel("Age")
plt.ylabel('Frequency')
plt.show()
```

Pie Chart

```python
s_count=df["Survived"].value_counts()
print(s_count)
```

```python
plt.figure(figsize=(8,6))
name=["Not Survived", "Survived"]
plt.pie(s_count, labels=name, autopct='%1.3f%%', startangle=90, colors=['purple', 'violet'])
plt.title("Survival distribution")
plt.show()
```

Scatter Plot

```python
plt.figure(figsize = (8,6))
plt.scatter(df['Age'], df['Fare'], c = 'skyblue', edgecolor = 'k', alpha = 0.5)
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

Line Chart

```python
plt.figure(figsize = (8,6))
plt.plot(df.index, df['Fare'], c = 'pink', lw = 2)
plt.title('Fare distribution')
plt.xlabel('Index')
plt.ylabel('Fare')
plt.show()
```

BoxPlot

```python
plt.figure(figsize = (8,6))
df.boxplot(column = 'Fare', by = 'Pclass', grid = False, patch_artist = True)
plt.suptitle('')
plt.title('Boxplot distribution')
plt.xlabel('P class')
plt.ylabel('Fare')
plt.show()
```

Donut Chart

```python
plt.figure(figsize = (8,6))
s_count = df["Survived"].value_counts()
name = ['Not Survived', 'Survived']
plt.pie(s_count, labels = name, autopct = '%1.3f%%', startangle = 90, colors = ['purple', 'violet'], wedgeprops=dict(width=0.4))
# autopct -> to display percentage with decimal 
# start angle -> to rotate the pie chart
plt.title("Survival distribution")
plt.show()
```
