```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("athletes.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.isnull().any()
```

```python
df.isnull().sum().sum()
```

```python
null_mean=df.isnull().mean()*100
print(null_mean)
```

```python
df.drop(columns=null_mean[null_mean>80].index, inplace=True)
```

```python
null_mean=df.isnull().mean()*100
print(null_mean)
```

```python
#mode is used for object 
object_col=df.select_dtypes(include=["object"]).columns
for col in object_col:
    mode_value=df[col].mode()[0]
    df[col].fillna(mode_value, inplace=True)
```

```python
#median is used for flat datatype
float_col=df.select_dtypes(include=["float"]).columns
for col in float_col:
    median_value=df[col].median()
    df[col].fillna(median_value, inplace=True)
```

```python
null_mean=df.isnull().mean()*100
print(null_mean)
```

```python
print(f"Null values after cleaning is: \n{df.isnull().sum()}")
```

```python
df.to_csv("athletes_cleaned.csv", index=False)
print("Cleaned dataset is saved into the athletes_cleaned.csv file")
```

Data Visualisation

```python
#Bar Graph
plt.figure(figsize=(8,6))
gender_count=df["gender"].value_counts()
gender_count.plot(kind="bar", color="purple", edgecolor="black")
plt.xlabel("Gender")
plt.ylabel("No. of athletes")
plt.xticks(rotation=0)
plt.show()
```

```python
#histogram
plt.figure(figsize=(8,6))
plt.hist(df["height"], bins=20, color="green", edgecolor="black")
plt.title("Histogram for height distribution")
plt.xlabel("Height")
plt.ylabel("Frequency")
```

```python
#pie chart
gender_count=df["gender"].value_counts()
print(gender_count)
```

```python
plt.figure(figsize=(8,6))
#gender_count=df["gender"].value_counts()
name=["Male", "Female"]
plt.pie(gender_count, labels=name, autopct="%1.3f%%",startangle = 90, colors=["lightgrey", "lightblue"])
plt.title("Pie chart for gender distribution")
plt.show()
```

```python
#Scatter Plot
plt.figure(figsize = (8,6))
plt.scatter(df['height'], df['weight'], c = 'lightpink', edgecolor = 'k', alpha = 0.5)
plt.title('Scatter plot for height vs weight')
plt.xlabel('height')
plt.ylabel('weight')
plt.show()
#alpha=0.5 → Adjusts transparency (50%) to prevent overlapping points from hiding others.
```

```python
#Line Chart
plt.figure(figsize = (8,6))
plt.plot(df.index, df['weight'], c = 'darkblue', lw = 2)
# lw -> line width
plt.title('Line Chart for weight distribution')
plt.xlabel('Athlete ID')
plt.ylabel('Weight')
plt.show()
```

```python
#Box Plot
plt.figure(figsize = (8,6))
df.boxplot(column = 'height', by = 'gender', grid = False, patch_artist = True)
plt.suptitle('')
plt.title('Boxplot distribution')
plt.xlabel('gender')
plt.ylabel('height')
plt.show()
```

```python
#Donut Chart
plt.figure(figsize = (8,6))
gender_count = df["gender"].value_counts()
name = ['Male', 'Female']
plt.pie(gender_count, labels = name, autopct = '%1.3f%%', startangle = 90, colors = ["lightgrey", "lightblue"], wedgeprops=dict(width=0.4))
# autopct -> to display percentage with decimal 
# start angle -> to rotate the pie chart
plt.title("Donut Chart for gender distribution")
plt.show()
```
